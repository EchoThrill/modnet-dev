```python
import pandas as pd

# Load the data
filepath = 'data/df_outputs_filtout.pkl'
df: pd.DataFrame = pd.read_pickle(filepath)
```

```python
df.head()
```

```python
from pymatgen.core.structure import Structure

# Extract the structure dict object from the dataframe
# It is a pymatgen.core.structure.Structure object encoded as a dict into a pickle file
df['structure'] = df['structure'].map(lambda x: Structure.from_dict(x))
df.head()
```

```python
from modnet.featurizers.presets import Matminer2024FastFeaturizer
from modnet.preprocessing import MODData

# Create a ModData object with the 2024 fast featurizer
data = MODData(materials=df['structure'],
               targets=df[['src_bandgap', 'src_n']],
               target_names=["Bandgap", "Refractive index"],
               structure_ids=df.index,
               featurizer=Matminer2024FastFeaturizer(),
               )
```

```python
data.featurize()
```

```python
data.get_featurized_df().head()
```

```python
data.save('out/featurized_data.pkl')
```
